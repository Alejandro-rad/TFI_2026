```python
import networkx as nx

GD = nx.DiGraph()
GD.add_nodes_from([1,2,3,4,5,6,7,8])
GD.add_edges_from([(1,2),(1,3),(2,3),(2,4),(3,4),(3,5),(4,5),(4,6),(5,6),(5,7),(6,7),(6,8),(7,8),(7,1),(8,1),(8,2)])

print("Los nodos de GD son: {}".format(list(GD.nodes)))
print("Las conexiones de GD son: {}".format(list(GD.edges)))

nx.draw(GD,with_labels=True,font_color='white',font_weight='bold',node_size=1000,pos=nx.circular_layout(GD))
```
